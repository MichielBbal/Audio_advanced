## 3. Creating tar files for the set of 25 images

source: https://www.tutorialspoint.com/How-to-create-a-tar-file-using-Python

```python
!cd /Users/michielbontenbal/Desktop/urban_sounds_small && ls
```

```python
#source: chatgpt
import os
import tarfile

def create_tar(folder_path, output_path):
    # Checking if the folder exists
    if not os.path.exists(folder_path):
        print(f"The folder '{folder_path}' does not exist.")
        return
    
    # Checking if the output path ends with '.tar'
    if not output_path.lower().endswith('.tar'):
        output_path += '.tar'

    with tarfile.open(output_path, 'w') as tar:
        # Walk through the directory and add each file to the tar file
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, folder_path)
                tar.add(file_path, arcname=arcname)

    print(f"Tar file created successfully at: {output_path}")

# Example usage:
#folder_path = '/Users/michielbontenbal/Desktop/urban_sounds_small'
output_path = 'urban_sounds_small.tar'

create_tar(folder_path, output_path)
```

```python
import tarfile

def read_tar(tar_path):
    try:
        with tarfile.open(tar_path, 'r') as tar:
            # Displaying the contents of the tar file
            for member in tar.getmembers():
                print(member.name)
    except tarfile.TarError as e:
        print(f"Error reading the tar file: {e}")

# Example usage:
tar_path = 'urban_sounds_small.tar'

read_tar(tar_path)
```
